# 1. 순환 이진 탐색

```python
def binary_search(a, left, right, K):
  if right >= left:
    mid = (left + right) //2

    if a[mid] == K:
      return True
    elif a[mid] > K:
      return binary_search(a, left, mid-1, K)
    else:
      return binary_search(a, mid+1, right, K)
  else:
    return False
```

```python
input_list = [10,29,25,35,45,55,60,75,85,90]
n = len(input_list)
```

```python
print('25탐색 : ', binary_search(input_list, 0, n-1, 25))
```

# 2. 스택

```python
stack = []  #스택 리스트 선언

def push(item):
  stack.append(item)

def pop():
  if len(stack) != 0:
    item = stack.pop(-1)
    return item
```

# [실습과제] 회문(Palindrome) 검사  
‘Anna’, ‘kayak’, ‘WoW’ 등과 같이 앞에서부터 읽거나 뒤에서부터 읽어도 같은 단어나 숫자, 문구 등을 팰린드롬(Palindrome)이라고 한다.  

입력한 문자열이 팰린드롬이면 ‘Yes’와 처음부터 중간까지의 문자를 출력한다.  

팰린드롬이 아니면 ‘No’와 원래의 문자열에 순서를 뒤집은 문자열을 연결하여 만든 팰린드롬을 출력하는 프로그램을 작성하시오.

[입력]  
알파벳으로 구성된 30자 이하의 문자열이 띄어쓰기 없이 주어진다.

[출력]  
1.주어진 문자열이 팰린드롬이면 첫 행에는 ‘Yes’, 둘째 행에는 팰린드롬 문자열의 처음부터 중간까지의 문자를 출력한다.

2.주어진 문자열이 팰린드롬이 아니면 첫 행에는 ‘No’, 둘째 행에는 주어진 문자열에 순서를 뒤집은 문자열을 연결하여 만든 팰린드롬을 출력한다.


## 여러 답안 공유
- 스택을 이용하여
- 문자열이나 리스트를 pop하여 뒤집을 수 있다.
- 전체를 뒤집어도 되고
- 반만 뒤집어도 팰린드롬 판별에는 무리가 없다.  
  (중간에 하나만 있는 글자는 판별에 영향을 미치지 않는다.)
- 4번 항목에 스택을 사용하지 않은 답도 참고삼아 공유함

### 1.문자열을 스택처럼

```python
def isPalindrome(s):
    strStack = ""

    #입력받은 문자열의 마지막부터 pop해서 반만 새 스택에 push
    for i in range(len(s)//2):
        strStack+=s[len(s)-1-i]

    #스택에 가져온 문자열과 원래 문자열 앞부분 반을 비교함
    if strStack == s[:len(s)//2]:
        return True
    else:
        return False
```

```python
inputStr1 = input('30자 이하 문자열을 영문으로 입력 :').replace(' ', '').lower()

if isPalindrome(inputStr1):
    print('Yes')
    print(inputStr1[:len(inputStr1)//2])
else:
    print('No')
    print(inputStr1+inputStr1[::-1])
```

### 2.리스트 스택
- 수업시간에 만든 리스트 스택을 사용
- 따라서 위에서부터 셀을 실행해야 아래의 셀들이 에러없이 실행됨

```python
# 리스트 스택을 사용
def isPalindrome2(strList):
  reverseHalf = []
  # pop()으로 문자열의 반을 가져온다. 홀수이면 남는 하나는 앞 리스트에 있게 됨
  for w in strList[:len(strList)//2]:
    reverseHalf.append(strList.pop())

  if strList[:len(reverseHalf)] == reverseHalf:
    return True
  else:
    return False
```

```python
inputStr2 = input('30자 이하 문자열을 영문으로 입력 :').replace(' ', '').lower()
if isPalindrome2(list(inputStr2)):
  print('Yes')
  print(inputStr2[:len(inputStr2)//2])
else:
  print('No')
  print(inputStr2+inputStr2[::-1])
```

### 3.스택클래스 직접 만들어 사용하기
- 스택 클래스를 만들고
- push(item), pop(), is_empty(), is_palindrome(s) 메소드를 만듬

```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        else:
            return None

    def is_empty(self):
        return len(self.items) == 0

def is_palindrome(s):
    s = s.replace(" ", "")
    s = s.lower()
    stack = Stack()

    for char in s[:len(s) // 2]:
        stack.push(char)

    for char in s[(len(s) + 1) // 2:]:
        if char != stack.pop():
            return False

    return True
```

```python
input_str = input()
if is_palindrome(input_str):
    print('Yes', input_str[:len(input_str)//2])
else:
    reversed_str = input_str[::-1]
    palindrome_str = input_str + reversed_str
    print('No', palindrome_str)
```

### 4.기타 여러가지 풀이
- 리스트 또는 문자열에서 첫 문자와 마지막 문자부터 차례로 짝을 지어 비교하여 판별
- 문자열이나 리스트를 한방에 뒤집는 기능을 사용하여 판별

```python
# 리스트 또는 문자열의 첫 인덱스와 마지막 인덱스의 값부터 순차로 비교하며
# 값이 다르면 빠져나온다.

stack = []
stack = list(input())
think = True  #
for i in range(0,len(stack) -1) :
    if stack[i] != stack[len(stack) -1 -i] :
        think = False
        break

if think :
    print("YES")
    print(stack[0: len(stack) // 2])
else :
    print("NO")
    stack.reverse()
    print(stack)
```

```python
# 문자열을 뒤집어 슬라이싱하거나 리스트의 reverse()를 이용하면 쉽게 할수 있다.
# 이렇게 해결한 사람은 스택 사용도 한 번 해 보자.

def is_pallindrome(s) :
    return s == s[::-1]

input_string = input()
if is_pallindrome(input_string) :
    print("yes")
    middle_index = len(input_string) // 2
    print(input_string[:middle_index])

else:
    print("no")
    reversed_string = input_string[::-1]
    pallindrome_string = input_string + reversed_string
    print(pallindrome_string)
```
